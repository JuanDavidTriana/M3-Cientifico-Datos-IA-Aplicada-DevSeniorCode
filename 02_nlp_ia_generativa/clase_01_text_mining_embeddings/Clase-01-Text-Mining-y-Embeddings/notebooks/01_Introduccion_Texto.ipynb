# Notebook 01 - Introduccion al Texto

## Objetivo
Comprender por que una computadora no entiende texto como lo hace una persona y como podemos analizar un texto paso a paso.

```python
from IPython.display import display
import pandas as pd

print("Entorno listo para iniciar el notebook 01")
```

## 1) Texto como string
En Python, una frase es una cadena de caracteres (`str`). Para la computadora, primero es una secuencia de simbolos, no significado.

```python
frase = "Me gusta aprender NLP con ejemplos practicos"
print("Frase original:", frase)
print("Tipo de dato:", type(frase))
print("Longitud total (caracteres):", len(frase))
```

## 2) Recorrido caracter por caracter
La computadora puede iterar cada simbolo individual para contar, transformar o filtrar texto.

```python
print("Primeros 20 caracteres con su posicion:")
for i, ch in enumerate(frase[:20]):
    print(f"indice={i:02d} -> '{ch}'")  # Muestra posicion y caracter
```

## 3) Separar palabras con split()
El metodo `split()` divide una cadena en una lista de palabras usando espacios como separador por defecto.

```python
palabras = frase.split()
print("Lista de palabras:", palabras)
print("Cantidad de palabras:", len(palabras))

for idx, palabra in enumerate(palabras, start=1):
    print(f"Palabra {idx}: {palabra}")
```

## 4) Diferencia entre caracteres, palabras y frases
En NLP trabajamos con distintos niveles de granularidad segun el problema.

```python
frases = [
    "La IA aplicada crea valor de negocio",
    "TensorFlow permite crear modelos de texto",
    "Los embeddings capturan similitud semantica"
]

resumen = []
for f in frases:
    tokens = f.split()
    resumen.append({
        "frase": f,
        "num_caracteres": len(f),
        "num_palabras": len(tokens),
    })

df_resumen = pd.DataFrame(resumen)
display(df_resumen)
```
