# Notebook 02 - Preprocesamiento

## Objetivo
Aplicar limpieza de texto paso a paso, mostrando el antes y despues de cada transformacion sin combinar todos los pasos de una sola vez.

```python
import re
import pandas as pd
from IPython.display import display

ruta = "../datasets/frases.csv"
df_base = pd.read_csv(ruta)

# Creamos un dataset pequeno con ruido para practicar limpieza paso a paso
df = df_base.head(10).copy()
df["texto_original"] = [
    f"{txt} !!! {2020 + i}  " if i % 2 == 0 else f"{txt} ???   {300 + i}"
    for i, txt in enumerate(df["frase"])
]
df = df[["texto_original"]]
display(df.head())
```

## 1) Convertir a minusculas
Reducimos variaciones artificiales entre palabras como `NLP` y `nlp`.

```python
df["paso_1_minusculas"] = df["texto_original"].str.lower()
display(df[["texto_original", "paso_1_minusculas"]].head())
```

## 2) Eliminar signos de puntuacion y simbolos
Usamos expresiones regulares para conservar letras, numeros y espacios.

```python
df["paso_2_sin_signos"] = df["paso_1_minusculas"].str.replace(r"[^\w\s]", " ", regex=True)
display(df[["paso_1_minusculas", "paso_2_sin_signos"]].head())
```

## 3) Eliminar numeros
Este paso depende del problema. En este ejemplo lo aplicamos para concentrarnos en vocabulario textual.

```python
df["paso_3_sin_numeros"] = df["paso_2_sin_signos"].str.replace(r"\d+", " ", regex=True)
display(df[["paso_2_sin_signos", "paso_3_sin_numeros"]].head())
```

## 4) Eliminar espacios duplicados
Despues de quitar signos y numeros suelen quedar multiples espacios.

```python
df["paso_4_espacios"] = (
    df["paso_3_sin_numeros"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
display(df[["paso_3_sin_numeros", "paso_4_espacios"]].head())
```

## Resultado final y comparacion
Observa el antes y despues completo para cada texto.

```python
display(df[["texto_original", "paso_4_espacios"]])

print("Total de filas procesadas:", len(df))
```
