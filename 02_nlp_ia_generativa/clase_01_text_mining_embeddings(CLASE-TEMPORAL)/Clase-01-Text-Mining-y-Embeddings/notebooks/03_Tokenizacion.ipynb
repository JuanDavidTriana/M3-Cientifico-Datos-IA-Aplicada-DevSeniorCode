# Notebook 03 - Tokenizacion

## Objetivo
Entender como `Tokenizer` transforma texto en indices numericos mediante `fit_on_texts()`, `word_index` y `texts_to_sequences()`.

## Flujo conceptual
texto -> palabras -> indices -> secuencias

Primero el tokenizador aprende el vocabulario, luego reemplaza cada palabra por su indice.

## Convertir texto a secuencias de enteros
Cada frase se transforma en una lista de indices enteros segun `word_index`.

## Revisar indices de palabras especificas
`word_index` nos permite inspeccionar como quedo codificado el vocabulario.
