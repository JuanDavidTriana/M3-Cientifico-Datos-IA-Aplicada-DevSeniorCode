# Notebook 03 - Tokenizacion

## Objetivo
Entender como `Tokenizer` transforma texto en indices numericos mediante `fit_on_texts()`, `word_index` y `texts_to_sequences()`.

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer

df = pd.read_csv("../datasets/frases.csv")
textos = df["frase"].head(12).tolist()

print("Numero de textos:", len(textos))
for i, t in enumerate(textos):
    print(f"Texto {i}: {t}")
```

## Flujo conceptual
texto -> palabras -> indices -> secuencias

Primero el tokenizador aprende el vocabulario, luego reemplaza cada palabra por su indice.

```python
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")

tokenizer.fit_on_texts(textos)

word_index = tokenizer.word_index

print("Tamaño word_index:", len(word_index))
for i, (palabra, indice) in enumerate(word_index.items()):
    print(f"Palabra: {palabra} -> Indice: {indice}")
```

## Convertir texto a secuencias de enteros
Cada frase se transforma en una lista de indices enteros segun `word_index`.

```python
secuencias = tokenizer.texts_to_sequences(textos)

for i in range(len(textos)):
    print(('-' * 40 ))
    print(f"Texto original: {textos[i]}")
    print(f"palabras: {textos[i].split()}")
    print(f"Secuencia: {secuencias[i]}")
    print(('-' * 40 ))
```

## Revisar indices de palabras especificas
`word_index` nos permite inspeccionar como quedo codificado el vocabulario.

```python
palabra_obetivo = ["nlp", "ia", "inteligencia", "artificial", "aprendizaje", "profundo"]
for p in palabra_obetivo:
    indice = word_index.get(p, None)
    if indice is not None:
        print(f"Palabra: {p} -> Indice: {indice}")
    else:
        print(f"Palabra: {p} no encontrada en el word_index.")
```
