## Objetivo
Entender por que las secuencias deben tener la misma longitud y como usar `pad_sequences()` en modo `pre` y `post`.

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

textos = [
    "nlp aplicado",
    "me gusta aprender embeddings",
    "esta clase explica padding y tokenizacion paso a paso",
    "ia"
]

tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(textos)
secuencias = tokenizer.texts_to_sequences(textos)

for t, s in zip(textos, secuencias):
    print(f"Texto: {t}")
    print(f"Secuencia: {s}")
    print()
```

## Por que igualar longitudes
Las redes neuronales procesan lotes (batches) de tensores rectangulares.
Si las secuencias tienen longitudes diferentes, necesitamos truncar o rellenar.

## Comparacion tabular
Mostramos secuencias originales y resultado de padding para facilitar la lectura.

## Cuando usar `pre` o `post`
- `padding='post'`: util cuando deseas preservar el inicio de la frase y completar al final.
- `padding='pre'`: comun en algunos modelos recurrentes antiguos donde importa alinear al final.

En la practica moderna con muchas arquitecturas, `post` suele ser mas intuitivo para texto natural.
